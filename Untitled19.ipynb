```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
```

```python
# 1. Base de Dados: [Regência, Estabilizador, Prumo, Escala]
# Alvo (y): 1 para À (crase), 0 para A (sem crase)
# Representação: [3, 0, 5, 2] -> Interpretados com características (features)

X_train = torch.tensor ([
    [1.0, 0.0, 1.0, 1.0], # Vou à escola (Pede 'a' + feminina + especifica) -> 1
    [1.0, 0.0, 0.0, 1.0], # Vou a Brasilia (pede 'a' + neutra) -> 0
    [0.0, 0.0, 1.0, 1.0], # Vi a escola (não pede 'a') ->
    [1.0, 0.0, 1.0, 0.0], # Vou a casa (Generico, sem especificação) ->
    [1.0, 0.0, 1.0, 1.0], # Vou àquela praça (Fusão com pronome) -> 1
], dtype=torch.float32)

y_train = torch.tensor([[1], [0], [0], [0], [1]], dtype=torch.float32)
```

```python
class ModeloCrase(nn.Module):
    def __init__(self):
        super(ModeloCrase, self).__init__()
        # Camada linear que funciona como nossa matriz de tranformação 3X3 + W
        self.volante = nn.Linear(4, 1)
        self.ativacao = nn.Sigmoid() # Convergenicia para 0 ou 1 (Solo)

    def forward(self, x):
        # A multiplicação da matriz pelo seu vetor [3, 0, 5, 2]
        retorno = self.volante(x)
        return self.ativacao(retorno)


model = ModeloCrase()

```

```python
# Critério de erro: Minimos Quadrados
criterion = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.1)

# Ciclo de Treino: O "Arremesso" repetido para aprender a trajetória
for epoch in range(1000):
    optimizer.zero_grad()
    outputs = model(X_train)
    loss = criterion(outputs, y_train) # Calcula a distância da curva ao solo real
    loss.backward() # Propaga a deformação toplológica (gradiente)
    optimizer.step() # Ajusta o volante (pesos)

print(f"Treinamento concluído em 2025. perda Final: {loss.item():.4f}]")
```

```python
# Teste: "Entreguei a encomenda à diretora"
# Regência: 1, Estabilizador: 0, Prumo Feminino: 1, Específico: 1
novo_arremesso = torch.tensor ([[1.0, 0.0, 1.0, 1.0]])
previsao = model(novo_arremesso)

print(f"Probabilidade de Crase: {previsao.item()*100:.2f}%")
if previsao.item() > 0.5:
    print("Resultado: À (fusão detectada!)")
else:
    print("REsultado: A (Sem crase)")
```

```python
# Simulando o "Dicionario de Regência" do modelo (Base de 2025)
regencia_verbos = {"vou": 1, "entreguei": 1, "vi": 0, "amo": 0}
especificidade_obj = {"escola": 1, "casa": 0, "diretora": 1, "brasilia": 0 }

def vetorizar_frase(verbo, substantivo):
  # X(3): Regência / Y(0): neutro / Z(5): Gênero / w(2): Especificidade
  x = regencia_verbos.get(verbo.lower(), 0)
  y = 0.0
  z = 1.0 # Assumindo feminino para o teste de crase
  w = especificidade_obj.get(substantivo.lower(), 0)
  return torch.tensor([[x, y, z, float(w)]], dtype=torch.float32)
```

```python
def processar_arremesso_gramatical(frase):
  # Exemplo simples de frase: "vou a escola"
    palavras = frase.lower().split()
    verbo = palavras[0]
    substantivo = palavras[- 1]

    # Gerando o vetor [3, 0, 5, 2] do objeto
    vetor_input = vetorizar_frase(verbo, substantivo)

    # Capturando a saída (A convergencia ao solo)
    with torch.no_grad():
        probabilidade = model(vetor_input).item()

    # Decodificação da trajetória
    if probabilidade > 0.5:
        frase_corrigida = frase.replace("a", "à")
        status = "CRITICAL COLLISION: À (crase detectada)"

    else:
        frase_corrigida = frase
        status = "LINEAR PATH: A (Sem fusão)"

    return frase_corrigida, status

# TESTE REAL
minha_frase = "Vou a escola"
resultado, info = processar_arremesso_gramatical(minha_frase)

print(f"Original: {minha_frase}")
print(f"Saida do Modelo {resultado}")
print(f"Análise Topologica: [{info}]")
```
