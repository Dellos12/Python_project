```python
import os
import pandas as pd
import torch
import numpy as np
```

```python
# preparação: Criando o arquivo físico (seu código base)

os.makedirs(os.path.join('..', 'data'), exist_ok=True)
data_file = os.path.join('..', 'data', 'house_tiny.csv')
with open(data_file, 'w') as f:
    f.write('NumRooms,Alley,price,GardenArea,RoofType\n')
    f.write('NA,Pave,127500,50,Tile\n')
    f.write('2,NA,106000,NA,Slate\n')
    f.write('4,NA,178100,NA,NA\n')
```

```python

# 2. Carregar is dadis usando Pandas
data = pd.read_csv(data_file)
print("--- Dados Carregasos do CSV ---")
print(data)
```

```python
# 3. Excluir a coluna com  a maioria dos valores ausente (Requerimento 3)
# Indentificamos qual coluna tem o máximo de valores nulos
coluna_para_remover = data.isnull().sum().idxmax()
data_limpo = data.drop(columns=[coluna_para_remover])
print(f"\n--- Coluna '{coluna_para_remover}' removida por ter mais nulos ---")
print(data_limpo)
```

```python
# 4. Preparação para tensor (Tratamento de dados categóricos e nulos)
# Tensores exigem números.  Vamosd preencher nulos numérico com média
# e converter categorias em colunas nunéricas (One-Hot Enconding)
inputs = data_limpo.drop(columns=['price']) # Separando Previsore
outputs = data_limpo['price']
```

```python
# Preencher valores numéricos nulos com média na própia coluna
inputs = inputs.fillna(inputs.mean(numeric_only=True))
```

```python
# Converter variáveis de texto (ex: RoofType) em números (0 e 1)
inputs = pd.get_dummies(inputs, dummy_na=True)

print("\n--- Dados preparados para conversão (Apenas Números)")
print(inputs)
```

```python
# 5. Converted para formato Tensor (requerimeto 4)
# Usamos float32, que é padrão para modelos de Deep Learning

X = torch.tensor(inputs.values.astype(float))
y = torch.tensor(outputs.values.astype(float))

print("\n--- Resultado Final: Formato Tensor ---")
print(X)
print(y)


```
